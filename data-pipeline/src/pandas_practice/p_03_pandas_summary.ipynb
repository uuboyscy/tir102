```python
"""
select *
from table
where portno = '2998623'
and percent_tna > 1.8
"""
filter_2998623 = df["portno"] == "2998623"
filter_percent_tna_gt_2 = df["percent_tna"] > 1.8

df[
    filter_2998623 & filter_percent_tna_gt_2
]
```

```python
"""
select *
from table
where percent_tna > 2
or percent_tna < 1
"""

filter_percent_tna_gt_2 = df["percent_tna"] > 2
filter_percent_tna_lt_1 = df["percent_tna"] < 1

df[
    filter_percent_tna_gt_2 | filter_percent_tna_lt_1
]
```

```python
"""
select *
from table
where portno = '2998623'
and (
    percent_tna > 2 or percent_tna < 1
)
"""

df[
    filter_2998623 & (
        filter_percent_tna_gt_2 | filter_percent_tna_lt_1
    )
]
```
